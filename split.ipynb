```python
import pandas as pd
import numpy as np
from src.article_predict import split_sentence
```

```python
file_path = 'data/article/A2_二次標註_網路危機訊息標註_1100927_僅全文標註(給致寧).xlsx'
df = pd.read_excel(file_path, sheet_name='contents')
```

```python
# filter out the rows with 'Crisis_Level' =='a:A'

A_df = df[df['Crisis_Level'] == 'a:A']
print(A_df.shape)
```

```python
article_key = 'Content(remove_tag)'
# iterate through the rows and split the sentences
# and save the sentences to a new dataframe

sentence_df = pd.DataFrame(columns=['TextID', 'Crisis_Level', 'Author', 'Title', 'Sentence'])

for index, row in A_df.iterrows():
    TextID = row['TextID']
    Crisis_Level = row['Crisis_Level']
    Author = row['Author'] 
    Title = row['Title']
    content = row[article_key]
    sentences = [sent for sent in split_sentence(content) if sent not in ['', ' ', '。', '，',',', '\n']]
    for sentence in sentences:
        sentence_df = sentence_df.append({'TextID': TextID, 'Crisis_Level': Crisis_Level, 'Author': Author, 'Title': Title, 'Sentence': sentence}, ignore_index=True)
```

```python
# save the sentence_df to a new excel file
print(sentence_df.shape)
sentence_df.to_excel('data/article/A2_A_sentences.xlsx', index=False)
```
